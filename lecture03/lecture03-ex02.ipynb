## An intuitive explanation what is DFT

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Signal
N = 8
signal = [1, 0, -1, 0, 1, 0, -1, 0]

# visualizing signal
plt.figure(figsize=(12, 8))
plt.scatter(np.arange(N), signal, label="observed signal", color = "red", s = 50)
plt.xlabel("Sample Index")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()
```

Let's calculate FFT of the signal.

```python
signal_fft = np.fft.fft(signal)
signal_fft
```

- a signal $\mathbf{f} = (1, 0, -1, 0, 1, 0, -1, 0)^\prime$.
- its DFT  $\widehat{\mathbf{f}} = (0, 0, 4, 0, 0, 0, 4, 0)^\prime$.
  + corresponds to frequencies $\frac{2}{8}$ and $-\frac{2}{8}$ (due to 
  **symmetry property** of real-valued signals, i.e., 
  $\widehat{\mathbf{f}}_k = \widehat{\mathbf{f}}_{N-k}$).
  + matches a cosine function $f(n) = \cos\left(\frac{2\pi\cdot 2n}{8}\right)=\cos\left(\frac{\pi n}{2}\right)$.
  + amplitude $A = \frac{2|\widehat{\mathbf{f}_k}|}{N}=\frac{2\cdot 4}{8} = 1$.
  + power is concentrated entirely at $\frac{1}{4}$, with no noise (other DFT bins are zero).

```python
# adding cos function cos(pi*n/2)

t = np.linspace(0, 7, 100)
f = np.cos(np.pi/2*t)

plt.figure(figsize=(12, 8))
plt.scatter(np.arange(N), signal, label="observed signal", color = "red", s = 50)
plt.plot(t, f, label="cosine wave", color = "green")
plt.xlabel("Sample Index")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()
plt.savefig("figures/lecture03-ex02.png", dpi = 300)
```
