# Lorenz dynamics

Loading necessary packages.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

Defining a `Python` function, representing the Lorents system.

```python
sigma, rho, beta= 10, 28, 8/3
theta = (sigma, rho, beta)

def lorenz(state, t0, theta = theta):
    x, y, z = state
    sigma, rho, beta = theta

    return [sigma*(y-x), x*(rho-z)-y, x*y-beta*z]
```

Simulating the dynamics, given a set of pre-specified parameters' values.

```python
state0 = (0, 1, 20)
dt = 0.001
t = np.arange(0, 50+dt, dt)

state = sp.integrate.odeint(lorenz, state0, t, rtol=10**(-12), atol=10**(-12)*np.ones_like(state0))
x, y, z = state.T
```

Visualization of the system trajectories.

```python
# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, 'b', lw=2)

# Optional: Add axis labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```
